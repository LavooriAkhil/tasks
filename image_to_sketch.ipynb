# Converting the Image to Sketch.


```python
def img_to_sketch(input_image):
  img= cv2.imread(input_image)
  print("original image: ")
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.axis('off')
  plt.imshow(img)
  plt.show()

  grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  inversion = cv2.bitwise_not(grey_img)

  blur_img = cv2.GaussianBlur(inversion,(21,21),0)

  inv_blur_img = cv2.bitwise_not(blur_img)

  pencil_img = cv2.divide(grey_img,inv_blur_img,scale = 256.0)
  print("Sketch of given image: ")
  plt.axis('off')
  plt.imshow(pencil_img, cmap= 'gray')
  plt.show()

```

```python
input_image = "/content/op.jpg"

img_to_sketch(input_image)
```
